```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://denison.cafebonappetit.com/cafe/2023-03-31/"
```

```python
res = requests.get(url)
res.raise_for_status()
print("code :" , res.status_code)
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
print(soup.title.get_text())
```

```python
# print(soup.a)
# print(soup.a.attrs)
# print(soup.a["href"])
```

```python
print(soup.find("main", attrs={"class" : "h4 site-panel__daypart-item-title"}))
```

```python
test = soup.find(attrs={"class" : "h4 site-panel__daypart-item-title"})
#print(test)
print(test.get_text())
```

```python
test1= soup.find(id = "breakfast")
test2 = test1.find(attrs={"class" :"c-tab__content site-panel__daypart-tab-content c-tab__content--active"})
test3 = test2.find_all(attrs={"class" : ["h4 site-panel__daypart-item-title", "site-panel__daypart-station-title"]}  )
flag_curtis = True
curtis = []
huffman = []
for i in range(len(test3)):
  if(test3[i].get_text() == "Huffman Café"):
    flag_curtis = False
  if(test3[i].get_text() == "Huffman Café" or test3[i].get_text() == "Curtis Café"):
    continue
  if(flag_curtis):
    curtis.append(test3[i].get_text().strip())
  else:
    huffman.append(test3[i].get_text().strip())

print(curtis)
print(huffman)
```
